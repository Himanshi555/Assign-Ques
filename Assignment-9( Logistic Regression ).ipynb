Question 2
Remove NaN values from the column ‘Survived’. 

```python
data['Survived'] =  data['Survived'].dropna()
data['Survived'].value_counts()
```

Question 3
Create a new column 'Has_Cabin' that tell us whether passengers had a cabin or not. 

```python
data['Has_Cabin'] = ~data.Cabin.isnull()
data.head(5)
```

Question 4
Create a new column “Title”, which will contain the titles of the passengers such as Mr., Mrs., etc

```python
data['Title'] = data.Name.apply(lambda x: re.search(' ([A-Z][a-z]+)\.', x).group(1))
data['Title'] = data['Title'].replace({'Mlle':'Miss', 'Mme':'Mrs', 'Ms':'Miss'})
data['Title'] = data['Title'].replace(['Rev', 'Dr',
                                            'Major', 'Sir'],'Special')
```

```python
data.head()
```

Question 5
Impute the missing values using .fillna() method.

```python
data['Age'] = data.Age.fillna(data.Age.median())
data['Fare'] = data.Fare.fillna(data.Fare.median())
data['Embarked'] = data['Embarked'].fillna('S')

data.head(5)
```

Question 6
Create two new columns “Categorical_age” and “Categorical_fare”, by using the pandas function qcut() 
to bin your numerical data.

```python
data['Categorical_age'] = pd.qcut(data.Age, q=4, labels=False )
data['Categorical_fare']= pd.qcut(data.Fare, q=4, labels=False)
data.head()
```

Question 7
Drop columns Cabin, Name, Passengerid, Ticket, Age, Fare, SibSp, Parch.

```python
data.drop(['Cabin', 'Name', 'PassengerId', 'Ticket', 'Age', 'Fare', 'SibSp', 'Parch'], axis=1, inplace=True)
data.head
```

Question 8
Transform all variables into numeric ones using get_dummies().

```python
df = pd.get_dummies(data, drop_first=True)
df.head()
```

Question 9
Now we need training set and testing set. Use train_test_split() to split your dataset. 
X(Features) will be the dataset values without the column “Survived” and y(Targets) will be “Survived” column values.

```python
X = df.drop(['Survived'], axis = 1).values
y = df['Survived'].values
```

```python
#Splitting the model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

Question 10
Use logistic regression by passing your X_train and y_train.

```python
from sklearn.linear_model import LogisticRegression
model =LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
test_pred = model.predict(X_test)
```

```python
accuracy_score = model.score(X_test, y_test)
print("The accuracy score is : {}".format(accuracy_score))
```
